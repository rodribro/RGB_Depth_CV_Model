```python
# Change RGB names to new defined convention to match the depth names (INESC Dataset)
def change_name(png_file_name):
    
    if png_file_name.startswith('RNGR'):
        png_file_name = png_file_name.replace('RNGR', '75G')

    if png_file_name.startswith('RWGR'):
        png_file_name = png_file_name.replace('RWGR', '50G')

    if png_file_name.startswith('RNRO'):
        png_file_name = png_file_name.replace('RNRO', '75R')
    
    return png_file_name
```

```python
from os import rename, listdir


def rename_rgb_dir(dir_folder):

    filenames = listdir(dir_folder)

    for png_file_name in filenames:

        old_path = (dir_folder + '/' + png_file_name)

        if png_file_name.startswith('RNGR'):
            new_name = png_file_name.replace('RNGR', '75G')
            new_path = (dir_folder + '/' + new_name)
            rename(old_path, new_path)

        elif png_file_name.startswith('RWGR'):
            new_name = png_file_name.replace('RWGR', '50G')
            new_path = (dir_folder + '/' + new_name)
            rename(old_path, new_path)

        elif png_file_name.startswith('RNRO'):
            new_name = png_file_name.replace('RNRO', '75R')
            new_path = (dir_folder + '/' + new_name)
            rename(old_path, new_path)

    print("Directory files successfully renamed!")
```

```python
import json
import os
```

```python
def rename_coordinate_dictionary_keys(json_file):

    with open(json_file) as c:
        coords = json.load(c)
    #print(coords)

    for item in coords:
        og_image_path = item["image_path"]
        
        #print(f"Original path: {og_image_path}")

        if og_image_path.startswith('RNGR'):
            new_name = og_image_path.replace('RNGR', '75G')
            item["image_path"] = new_name

        elif og_image_path.startswith('RWGR'):
            new_name = og_image_path.replace('RWGR', '50G')
            item["image_path"] = new_name

        elif og_image_path.startswith('RNRO'):
            new_name = og_image_path.replace('RNRO', '75R')
            item["image_path"] = new_name

        print(f"Original path: {item["image_path"]}")

    with open("renamed_INESC_coords.json", "w") as nc:
        json.dump(coords, nc)

rename_coordinate_dictionary_keys("../inesc_coordinates.json")
```
